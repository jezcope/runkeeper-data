```python
cardioActivities = pd.read_csv('data/cardioActivities.csv', parse_dates=[0, 4, 5], index_col=0)
display(cardioActivities.head())
```

```python
cardioActivities['Type'] = cardioActivities['Type'].astype('category')
print(cardioActivities['Type'].cat.categories)
```

```python
runActivities = cardioActivities[cardioActivities['Type'] == 'Running']
runActivities['Distance (mi)'].plot()
```

```python
runActivities = cardioActivities[(cardioActivities['Type'] == 'Running') & cardioActivities['Distance (mi)'].notnull()]
runActivities['Distance (mi)'].plot()
```

```python
runActivities = runActivities[(runActivities['Distance (mi)'] <= 15)
                             & runActivities['Average Speed (mph)'].between(3.5, 10)]
#runActivities = runActivities.loc[runActivities.index > '2015-01-01']
runActivities['Distance (mi)'].plot()
```
